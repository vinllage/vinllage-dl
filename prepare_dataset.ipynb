```python
import json

API1 = "MlZ9K6KsGHXzBfZyXfP8"  # rf1
API2 = "JvPKQh62pMnrK9yNltri"  # rf2

def list_projects(api_key, workspace_slug):
    url = f"https://api.roboflow.com/{workspace_slug}?api_key={api_key}"
    r = requests.get(url)
    j = r.json()
    print(f"\n[{workspace_slug}] status={r.status_code}")
    print(json.dumps(j, indent=2, ensure_ascii=False))
    # 로보플로우 응답 구조가 업데이트될 수 있어, 안전하게 필드 탐색
    # 흔히 'workspace' 아래 'projects' 배열에 각 프로젝트의 'slug'가 들어있습니다.
    try:
        projects = j.get("workspace", {}).get("projects", [])
        print("\n프로젝트 슬러그들:")
        for p in projects:
            # p 예: {"id":"...", "name":"...", "slug":"my-project", ...}
            print("-", p.get("slug"), "| name:", p.get("name"))
    except Exception as e:
        print("프로젝트 필드 파싱 실패:", e)

# 두 워크스페이스에 대해 실행
list_projects(API1, "daeun-jiao6")
list_projects(API2, "team2-hualr")
```

```python
# Roboflow에서 두 데이터셋 내려받기

from roboflow import Roboflow
from pathlib import Path

API_A = "MlZ9K6KsGHXzBfZyXfP8"    # daeun-jiao6 접근 권한 키
API_B = "JvPKQh62pMnrK9yNltri"    # team2-hualr 접근 권한 키
VER_A = 2                           # A 프로젝트 사용할 버전
VER_B = 1                           # B 프로젝트 사용할 버전

# 슬러그(확정)
WS_A, PRJ_A = "daeun-jiao6", "team2-xfdnc"
WS_B, PRJ_B = "team2-hualr", "team2-dsmnt"

rfA = Roboflow(api_key=API_A)
rfB = Roboflow(api_key=API_B)

rootA = Path(rfA.workspace(WS_A).project(PRJ_A).version(VER_A).download("yolov8").location)
rootB = Path(rfB.workspace(WS_B).project(PRJ_B).version(VER_B).download("yolov8").location)

print("A:", rootA)
print("B:", rootB)
```

```python
import yaml
from pathlib import Path

# 1) 각 data.yaml에서 클래스 불러오기
namesA = yaml.safe_load(open(rootA/'data.yaml', encoding='utf-8'))['names']
namesB = yaml.safe_load(open(rootB/'data.yaml', encoding='utf-8'))['names']
print("A:", namesA)
print("B:", namesB)

# 2) 최종 클래스 배열 = A 순서 유지 + (B에만 있는 라벨) 뒤에 추가
final_names = namesA[:]  # ['can','glass','paper','plastic','recycle','sticker','vinyl']
name_to_idx_A = {n:i for i,n in enumerate(final_names)}

# A -> final
mapping_A_to_final = {i: name_to_idx_A[n] for i,n in enumerate(namesA)}

# B -> final (aluminum → can)
mapping_B_to_final = {}
for i,n in enumerate(namesB):
    if n == "aluminum":
        target = "can"
    else:
        target = n
    if target in name_to_idx_A:
        mapping_B_to_final[i] = name_to_idx_A[target]

print("FINAL:", final_names)
print("A->FINAL:", mapping_A_to_final)
print("B->FINAL:", mapping_B_to_final)
```

```python
# 레이블 리매핑(A/B → 최종 인덱스)

from pathlib import Path

def remap_yolo_labels(label_dir: Path, mapping: dict):
    if not label_dir.exists(): return
    for p in label_dir.glob("*.txt"):
        lines = p.read_text(encoding='utf-8').splitlines()
        out = []
        for line in lines:
            if not line.strip(): continue
            parts = line.split()
            cls = int(parts[0])
            if cls not in mapping: continue
            parts[0] = str(mapping[cls])
            out.append(" ".join(parts))
        p.write_text("\n".join(out), encoding='utf-8')

for split in ["train","valid","test"]:
    remap_yolo_labels(rootA/f"{split}/labels", mapping_A_to_final)
    remap_yolo_labels(rootB/f"{split}/labels", mapping_B_to_final)

print(">> labels remapped")
```

```python
# 물리 병합(파일명 충돌 방지) + 검증 10% 분리

import shutil, random
from tqdm import tqdm

merged = Path("merged_can_union")

def merge_split(src_root: Path, dst_root: Path, prefix: str):
    for split in ["train","valid","test"]:
        for sub in ["images","labels"]:
            (dst_root/split/sub).mkdir(parents=True, exist_ok=True)
            src_dir = src_root/split/sub
            if not src_dir.exists(): continue
            for p in tqdm(list(src_dir.glob("*")), desc=f"{prefix}:{split}/{sub}"):
                (dst_root/split/sub/f"{prefix}__{p.name}").write_bytes(p.read_bytes())

merge_split(rootA, merged, "A")
merge_split(rootB, merged, "B")
print("Merged at:", merged.resolve())

# valid 없으면 train에서 10% 이동
train_imgs = sorted((merged/"train/images").glob("*"))
vimg, vlbl = merged/"valid/images", merged/"valid/labels"
vimg.mkdir(parents=True, exist_ok=True); vlbl.mkdir(parents=True, exist_ok=True)

if len(list(vimg.glob("*"))) == 0 and len(train_imgs) > 0:
    n = max(1, int(0.1*len(train_imgs)))
    for img in random.sample(train_imgs, n):
        lbl = merged/"train/labels"/(img.stem + ".txt")
        shutil.move(str(img), str(vimg/img.name))
        if lbl.exists(): shutil.move(str(lbl), str(vlbl/lbl.name))
    print(f">> moved {n} images to valid/")
```

```python
# 최종 data.yaml 생성

import yaml
final_names = ['can','glass','paper','plastic','recycle','sticker','vinyl']  # A 기준

yaml.safe_dump({
    "path": str(merged),
    "train": str(merged/"train/images"),
    "val":   str(merged/"valid/images"),
    "test":  str(merged/"test/images"),
    "names": final_names
}, open("merged_dataset.yaml","w",encoding="utf-8"))

print(">> merged_dataset.yaml created")
```
